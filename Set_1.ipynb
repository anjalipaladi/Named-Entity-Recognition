```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
common = [x for x in list1 if x in list2]
print(common)
```

```python
sentence = "the cat and the hat"
words = sentence.split()
result = {word: words.count(word) for word in set(words)}
print(result)
```
