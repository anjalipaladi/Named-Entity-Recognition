```python
numbers = [4, 2, 4, 5, 2, 3, 5, 1]
result = []
for num in numbers:
    if num not in result:
        result.append(num)
print(result)
```

```python
lst = [2, 4, 3, 5, 7, 8, -1]
target = 6
pairs = set()
for num in lst:
    if (target - num) in lst:
        pairs.add(tuple(sorted((num, target - num))))
print(list(pairs))
```
